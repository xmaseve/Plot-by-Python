```python
from sklearn.datasets import load_iris
import pylab as plt
import numpy as np
%pylab inline
```

```python
iris = load_iris()
```

```python
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
for row in iris.data:
    sepal_length.append(row[0])
    sepal_width.append(row[1])
    petal_length.append(row[2])
    petal_width.append(row[3])
```

```python
plt.figure('sepal')
plt.clf()
plt.plot(sepal_length,sepal_width,'ro',label='Sepal')
plt.plot(petal_length,petal_width,'bo',label='Petal')
plt.legend(loc='upper left')    
plt.xlabel('length')
plt.ylabel('width')
plt.title('Sepal & Petal')
```

```python
plt.figure('hist')
plt.clf()
plt.subplot(211)
plt.hist(sepal_length,50,facecolor='g',label='Sepal length')
plt.legend()
plt.grid(True)
plt.subplot(212)
plt.hist(sepal_width,50,facecolor='r',label='Sepal width')
plt.legend()
plt.grid(True)
```

```python
length_data = [sepal_length, petal_length]
width_data = [sepal_width, petal_width]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# rectangular box plot
bplot1 = axes[0].boxplot(length_data,
                         vert=True,   # vertical box aligmnent
                         patch_artist=True)   # fill with color

# notch shape box plot
bplot2 = axes[1].boxplot(width_data,
                         notch=True,  # notch shape
                         vert=True,   # vertical box aligmnent
                         patch_artist=True)   # fill with color

# fill with colors
colors = ['pink', 'lightblue']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(length_data))], )
    ax.set_xlabel('xlabel')
    ax.set_ylabel('ylabel')

# add x-tick labels
plt.setp(axes, xticks=[y+1 for y in range(len(length_data))],
         xticklabels=["Sepal", "Petal"])
```
